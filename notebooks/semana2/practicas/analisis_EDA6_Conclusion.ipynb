## Conclusión General
El análisis demuestra que el principal factor que afecta la rentabilidad es el descuento aplicado.
El negocio no debe evaluarse únicamente por volumen de ventas, sino por margen y profit sostenido.
Las decisiones estratégicas deben enfocarse en:
- Control de descuentos.
- Optimización de categorías con profit negativo.
- Foco en segmentos más rentables como Home Office.
- Ajustes regionales según desempeño real.
